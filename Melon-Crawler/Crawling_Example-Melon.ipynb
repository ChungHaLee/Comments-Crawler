```python
# ë¶ˆìš©ì–´ ì‚¬ì „ ë¶ˆëŸ¬ì˜¤ê¸°
with open('stop_words.txt', 'r') as file:
    stop_word = file.readline()
    stop_word = str(stop_word)
    
stop_word = stop_word.replace("\ufeff", '').replace("'", '').replace(",", '').replace('\n', '').replace("â€™", '').replace("â€˜", '')
stop_words = stop_word.split()
```

```python
driver = wd.Chrome(executable_path='/Users/lifeofpy/Desktop/chromedriver')
url = 'https://www.melon.com/song/detail.htm?songId=34041583#cmtpgn=&pageNo={}&sortType=0&srchType=2&srchWord='
comments = []
```

```python
# í˜ì´ì§€ ê°œìˆ˜ ê°€ì ¸ì˜¤ê¸°
# í˜ì´ì§€ ê°œìˆ˜ë¥¼ ê°€ì ¸ì˜¤ê¸° ìœ„í•´ì„œëŠ” ëŒ“ê¸€ ê°œìˆ˜ê°€ í•„ìš”í•˜ë‹¤. 
# (í•˜ë‚˜ì˜ í˜ì´ì§€ì— ëŒ“ê¸€ 10ê°œê°€ ìˆê¸° ë•Œë¬¸ì— ëŒ“ê¸€ ê°œìˆ˜ ë‚˜ëˆ„ê¸° 10 = í˜ì´ì§€ ê°œìˆ˜)
page_num = 0
page_url = 'https://www.melon.com/song/detail.htm?songId=34041583'
driver.get(page_url)
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
pages = soup.find_all('span', {'class': 'd_cmtpgn_srch_cnt'})[0]

# ì¼ë‹¨ ì¶”ì¶œëœ íƒœê·¸ë¥¼ ë¬¸ìì—´ë¡œ ë°”ê¾¸ê¸°
pages = str(pages)

# ì •ê·œì‹ì„ í†µí•´ íƒœê·¸ì—ì„œ ìˆ«ì(=ëŒ“ê¸€ ê°œìˆ˜)ë§Œ ì¶”ì¶œí•˜ê¸°
cmt_num = re.findall('\d+', pages)
page_num = "".join(cmt_num)
page_num = int(page_num)
page_num = page_num//10

# ì¼ë‹¨ ì¶”ì¶œëœ íƒœê·¸ë¥¼ ë¬¸ìì—´ë¡œ ë°”ê¾¸ê¸°
pages = str(pages)

# ì •ê·œì‹ì„ í†µí•´ íƒœê·¸ì—ì„œ ìˆ«ì(=ëŒ“ê¸€ ê°œìˆ˜)ë§Œ ì¶”ì¶œí•˜ê¸°
cmt_num = re.findall('\d+', pages)

for i in cmt_num:
    page_num = int(i) // 10

# í˜ì´ì§€ ê°œìˆ˜
page_num
```

```python
# ë©œë¡  í™ˆí˜ì´ì§€ì—ì„œ ëŒ“ê¸€ ê°€ì ¸ì™€ì„œ ë¦¬ìŠ¤íŠ¸ë¡œ ì €ì¥í•˜ê¸°
for i in range(0, page_num+2):
    link = url.format(i)
    driver.get(link)
    driver.implicitly_wait(2)
    
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    comment = soup.find_all('div', {'class': 'd_cmtpgn_cmt_full_contents'})
    
    for r in comment:
        comments.append(r.get_text().strip())
```

```python
melon_comments = []
melon_comments = comments
```

```python
# ëŒ“ê¸€ì´ ëª¨ì¸ comments ë¦¬ìŠ¤íŠ¸ì—ì„œ í•„ìš”ì—†ëŠ” ë¬¸ì ì²˜ë¦¬í•˜ê¸°
for j in range(len(melon_comments)):
    if 'ë‚´ìš©' in melon_comments[j]:
        melon_comments[j] = melon_comments[j].replace('ë‚´ìš©', '')
```

```python
# íŠ¸ìœ„í„° í˜•íƒœì†Œ ë¶„ì„ê¸° Okt ë¥¼ ì‚¬ìš©
from konlpy.tag import Okt
okt = Okt()

# ëŒ“ê¸€ì„ ë¦¬ìŠ¤íŠ¸ë¡œ ë‹´ê¸° ìœ„í•´ okts ë¦¬ìŠ¤íŠ¸ ìƒì„±
okts = []
```

```python
for i in range(len(melon_comments)):
    okts.append(okt.pos(melon_comments[i]))
```

```python
okts
```

```python
adj = []

for i in range(len(okts)):
    for j in range(len(okts[i])):
        if okts[i][j][1] == 'Adjective' and len(okts[i][j][0]) > 1:
            adj.append(okts[i][j][0])
```

```python
adj = set(adj) # ë¦¬ìŠ¤íŠ¸ ì›ì†Œ ì¤‘ë³µ ì œê±°
print(adj)
```

```python
# ì§„ì§œ í˜•ìš©ì‚¬ë§Œ ëª¨ì•„ë†“ì„ real_adj ë¦¬ìŠ¤íŠ¸ ì •ì˜

adj = list(adj) # ë‹¤ì‹œ ì¤‘ë³µ ì œê±° ì—†ì• ê¸°

real_adj = []
for w in adj:
    if w not in stop_words:
        real_adj.append(w)
        
print(real_adj)
```
